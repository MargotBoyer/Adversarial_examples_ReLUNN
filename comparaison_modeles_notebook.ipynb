```python
import numpy as np
import pandas as pd
```

```python
ben = pd.read_csv("datasets\MNIST\Benchmark\MNIST_epsilon=1_neurones=34_taille=6_benchmark.csv\MNIST_epsilon=1_neurones=34_taille=6_benchmark.csv")
```

```python
ben.head()
```

```python
ben["REUSSI"] = (ben["y0"] != ben["Label"]) & (ben["Label"] != -1)
```

```python
ben_true_values = ben[(ben["modele_opt"] == "Fischetti_Obj_diff") & (ben["relax"] == 0)]
# Fonction qui retourne la vraie valeur optimale du problème non relâche
def renvoie_vraie_valeur_optimale(row, df):
    corresponding_row = ben_true_values[(ben_true_values['x0_id'] == row['x0_id'])]
    if not corresponding_row.empty:
        return corresponding_row.iloc[0]['Valeur_optimale']
    return None
```

```python
ben_true_values["Valeur_optimale"].head()
```

```python
# Création d'une nouvelle colonne contenant la vraie valeur optimale du problème pour 
ben['Valeur_optimale_ref'] = ben.apply(lambda row: renvoie_vraie_valeur_optimale(row, ben), axis=1)
```
